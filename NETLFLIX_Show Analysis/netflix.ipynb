```python
raw.drop(raw[raw.title.isnull()].index, inplace=True)
raw.drop(raw[raw.description.isnull()].index, inplace=True)
raw.drop(raw[raw.age_certification .isnull()].index, inplace=True)
raw.drop(raw[raw.seasons.isnull()].index, inplace=True)
raw.drop(raw[raw.imdb_id .isnull()].index, inplace=True)
raw.drop(raw[raw.imdb_score .isnull()].index, inplace=True)
raw.drop(raw[raw.imdb_votes.isnull()].index, inplace=True)
raw.drop(raw[raw.tmdb_popularity.isnull()].index, inplace=True)
raw.drop(raw[raw.tmdb_score.isnull()].index, inplace=True)
```
