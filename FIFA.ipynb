```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.size'] = 12
plt.figure(figsize=(15,6))
sns.barplot(x="Team",y="Win",data=read1.sort_values(["Team","Points"]).head(10))
plt.xticks(rotation=40)
plt.xlabel("TEAMS")
plt.ylabel("WINS")
plt.title("Data Visualisation")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.size'] = 14
plt.figure(figsize=(15,6))
sns.barplot(x="Team",y="Win",data=read1.sort_values(["Team","Win"]).tail(5))
plt.xticks(rotation=40)
plt.xlabel("TEAMS")
plt.ylabel("WINS")
plt.title("Data Visualisation")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.size'] = 14
plt.figure(figsize=(15,6))
sns.barplot(x="Team",y="Goals For",data=read1.sort_values(["Team","Goals For"]).head(5))
plt.xticks(rotation=40)
plt.xlabel("TEAMS")
plt.ylabel("GOALS")
plt.title("Data Visualisation")
```
