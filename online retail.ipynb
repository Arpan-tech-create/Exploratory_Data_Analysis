# Let's Start With Saudi Arabia Country

```python
new=raw_data[['StockCode','Description','UnitPrice']]
new_df=raw_data.Country=='Saudi Arabia'
```

```python
arabia=new[new_df]
arabia_df=arabia.sort_values(by=['UnitPrice'],ascending=False)
arabia_df
```

```python
plt.figure(figsize=(15,8))
sns.barplot(y='Description',x='UnitPrice',data=arabia_df,edgecolor='darkblue')
```

So, From Above Barplot , we can say that in Saudi Arabia , GOLD EAR MUFF HEADPHONES have Maximum Unitprice.

# About Singapore

```python
sau_cnt=raw_data[['StockCode','Description','UnitPrice']]
cnt_df=raw_data.Country=='Singapore'
```

```python
saudi=sau_cnt[cnt_df]
saudi_df=saudi.sort_values(by=['UnitPrice'],ascending=True).head(10)
saudi_df
```

```python
plt.xticks(rotation=90)
sns.barplot(y='Description',x='UnitPrice',data=saudi_df,palette='Blues_d',edgecolor='green')
```

# USA Country

```python
usa_cnt=raw_data[['StockCode','Description','UnitPrice','Quantity']]
usa_df=raw_data.Country=='USA'
```

```python
usa_num=usa_cnt[usa_df]
final_usa=usa_num.sort_values(by=['Quantity'],ascending=True).head(20)
final_usa
```

```python
name=final_usa.Description
unit=final_usa.UnitPrice
quantity=final_usa.Quantity
```

```python
plt.xticks(rotation=90)
plt.plot(name,unit,marker='o',color='purple',mfc='y',markersize=10)
plt.plot(name,quantity,marker='*',color='green',mfc='y',markersize=10)
plt.legend(['UnitPrice','Quantity'])
```
