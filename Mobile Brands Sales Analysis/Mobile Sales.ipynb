# Which One Is The Most Popular Processor ? 

```python
import math

def roundup(x):
    return 100 + int(math.ceil(x / 100.0)) * 100 

sns.factorplot('processor',data=read, kind='count', size=14, aspect=1,palette='Blues_d')
ax = plt.gca()
y_max = read['processor'].value_counts().max() 
ax.set_ylim([0, roundup(y_max)])
plt.xticks(rotation=60)
# Iterate through the list of axes' patches
for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2., p.get_height(), '%d' % int(p.get_height()), 
            fontsize=14, color='blue', ha='center', va='bottom')

plt.show()
```

```python
comment_words = ' '
  
for val in read.model.values: 
    val = str(val) 
    tokens = val.split() 
        
    for i in range(len(tokens)): 
        tokens[i] = tokens[i].lower() 
          
    for words in tokens: 
        comment_words = comment_words + words + ' '
  
  
wordcloud = WordCloud(width = 1000, height = 1000, 
                background_color ='black', 
                min_font_size = 10).generate(comment_words) 
wordcloud.to_image()
```
