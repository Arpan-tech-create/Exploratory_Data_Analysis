```python
plt.figure(figsize=(15,7))
raw_data['Activity'].value_counts().head(10).plot.pie(autopct='%1.1f%%')
plt.legend(raw_data['Activity'].value_counts(),loc='upper left')
```

```python
plt.figure(figsize=(15,7))
sns.distplot(raw_data['Debit'],color='r').set(title='Debits')
```

```python
plt.figure(figsize=(15,7))
sns.distplot(raw_data['Credit'],color='g').set(title='Credits')
```

```python
plt.figure(figsize=(15,7))
ax1 = plt.subplot()
ax2 = ax1.twinx()
sns.scatterplot(x='Activity',y='Debit',color='r',data=raw_data,ax=ax1)
sns.scatterplot(x='Activity',y='Credit',color='g',data=raw_data,ax=ax2)
ax2.tick_params(axis='y', colors='g')
ax1.tick_params(axis='x',rotation=90)
```

```python
val=raw_data[['Date','Source','Debit']]
val_df=raw_data.Activity=='Paid for order'
```

```python
fetch=val[val_df]
fetch_df=fetch.sort_values(by=['Debit'],ascending=True).head(10)
fetch_df
```

```python
fetch_df.Debit.sum()
```

```python
plt.figure(figsize=(15,7))
plt.xticks(rotation=80)
sns.barplot(x='Source',y='Debit',data=fetch_df,palette='magma')
```
