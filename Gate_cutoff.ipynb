```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


plt.rcParams['font.size'] = 15
sns.set_style('darkgrid')
```

```python
raw_data=pd.read_csv('gate_cutoff123.csv')
```

```python
raw_data
```

```python
raw_data['OBC'].max()
```

```python
raw_data.drop(['SrNo'],axis=1,inplace=True)
```

```python
raw_data
```

```python
raw_data.isnull().sum()
```

```python
raw_df=raw_data.sort_values(by=['Year'],ascending=False).head(20)
```

```python
raw_df
```

```python
top_paper=raw_df.PaperName
obc=raw_df.OBC
gen=raw_df.GEN
sc_st=raw_df.SC_ST
```

```python
plt.figure(figsize=(16,9))
plt.xticks(rotation=80)
plt.plot(top_paper,obc,color='purple',linewidth=5)
plt.plot(top_paper,gen,color='green',marker='o',mec='yellow',mfc='red',ms=8,linewidth=5)
plt.plot(top_paper,sc_st,color='blue',marker='o',mec='yellow',mfc='red',ms=8,linewidth=5)
plt.legend(['OBC','GEN','SC_ST']);
plt.xlabel('Paper_Name')
plt.ylabel('CASTS');
```

```python
comment_words = ' '
  
for val in raw_data.PaperName.values: 
    val = str(val) 
    tokens = val.split() 
        
    for i in range(len(tokens)): 
        tokens[i] = tokens[i].lower() 
          
    for words in tokens: 
        comment_words = comment_words + words + ' '
  
  
wordcloud = WordCloud(width = 1000, height = 1000, 
                background_color ='black', 
                min_font_size = 10).generate(comment_words) 
wordcloud.to_image()
```
