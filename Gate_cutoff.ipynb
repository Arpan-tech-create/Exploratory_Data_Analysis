```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


plt.rcParams['font.size'] = 13
plt.rcParams['figure.figsize'] = (9, 5)
plt.rcParams['figure.facecolor'] = '#00000000'
sns.set_style('darkgrid')
```

```python
raw_data=pd.read_csv('gate_cutoff123.csv')
```

```python
raw_data
```

```python
raw_data['OBC'].max()
```

```python
raw_data.drop(['SrNo'],axis=1,inplace=True)
```

```python
raw_data
```

```python
raw_data.isnull().sum()
```

```python
raw_df=raw_data.sort_values(by=['Year'],ascending=False).head(20)
```

```python
raw_df
```

```python
top_paper=raw_df.PaperName
obc=raw_df.OBC
gen=raw_df.GEN
sc_st=raw_df.SC_ST
```

```python
plt.figure(figsize=(15,8))
plt.xticks(rotation=80)
plt.plot(top_paper,obc,color='purple')
plt.plot(top_paper,gen,color='green')
plt.plot(top_paper,sc_st,color='red')
plt.legend(['OBC','GEN','SC_ST']);
plt.xlabel('Paper_Name')
plt.ylabel('CASTS');
```
