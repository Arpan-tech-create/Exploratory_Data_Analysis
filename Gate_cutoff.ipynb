```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


plt.rcParams['font.size'] = 13
plt.rcParams['figure.figsize'] = (9, 5)
plt.rcParams['figure.facecolor'] = '#00000000'
```

```python
raw_data=pd.read_csv('gate_cutoff123.csv')
```

```python
raw_data
```

```python
raw_data.drop(['Sr No'],axis=1,inplace=True)
```

```python
raw_data
```

```python
raw_data.isnull().sum()
```

# For General Category

```python
raw_paper=['PaperName','GEN','Paper Code']
paper_df=raw_data[raw_paper]
```

```python
paper_df
```

```python
plt.figure(figsize=(15,8))
plt.title("General category", color='r',size=15)
plt.xticks(rotation=0)
plt.grid()
sns.barplot(y='PaperName',x='GEN',data=paper_df)
```

```python
plt.figure(figsize=(15,8))
plt.title("General category", color='r',size=15)
sns.barplot(y='PaperName',x='GEN',data=paper_df.head(10))
```

```python
plt.figure(figsize=(15,8))
plt.title("General category", color='r',size=15)
plt.xticks(rotation=0)
plt.grid()
plt.scatter(y='PaperName',x='GEN',data=paper_df.head(30))
```

# For OBC Category

```python
obc_var=['PaperName','OBC']
OBC_df=raw_data[obc_var]
```

```python
OBC_df
```

```python
OBC_df.PaperName.value_counts()
```

```python
plt.figure(figsize=(15,8))
plt.title("OBC Category", color='r',size=15)
plt.grid(color='r')
sns.barplot(y='PaperName',x='OBC',palette="Blues_d",data=OBC_df.head(10))
```
