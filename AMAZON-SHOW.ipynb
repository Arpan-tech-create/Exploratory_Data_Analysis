Cleared all the duplicates values

```python
raw.isnull().sum()
```

```python
raw.drop(raw[raw.description.isnull()].index,inplace=True)
raw.drop(raw[raw.imdb_score.isnull()].index,inplace=True)
raw.drop(raw[raw.imdb_votes.isnull()].index,inplace=True)
raw.drop(raw[raw.tmdb_popularity.isnull()].index,inplace=True)
raw.drop(raw[raw.tmdb_score.isnull()].index,inplace=True)
```

all the null values have been removed

```python
raw.drop(raw[raw.release_year<=2000].index,inplace=True)
```

# BLock-BUSTER SHOW Based on IMDB-VOtes

```python
sorted=raw.sort_values(by=['imdb_votes'],ascending=False).head(10)
sorted
```

```python
raw.drop(['description'],axis=1,inplace=True)
```

```python
raw
```

```python
plt.figure(figsize=(15,8))
plt.xticks(rotation=80)
plt.title('Max IMDB Voting')
sns.barplot(x='title',y='imdb_votes',data=sorted,palette='dark')
```

# POPULAR GENRES

```python
gen=raw.genres.value_counts().head(10)
gen
```

```python
plt.figure(figsize=(15,7))
plt.title('Popular GENRES')
plt.grid()
plt.xticks(rotation=60)
sns.barplot(y=gen,x=gen.index,palette='rocket')
```

# Popular INDIAN Show OR Movie on AMAZON

```python
india=raw[['title','production_countries','genres','imdb_votes','release_year']]
new_in=raw.production_countries=='IN'
```

```python
new_con=india[new_in]
india_df=new_con.sort_values(by=['imdb_votes'],ascending=False).head(10)
india_df
```

```python
plt.figure(figsize=(15,7))
plt.title('Popular Indian Show OR Movie')
plt.grid()
plt.xticks(rotation=60)
sns.barplot(x='title',y='imdb_votes',palette='magma',data=india_df)
```
