```python
TITLE1=shows_df.title
im12=shows_df.imdb_votes
tm12=shows_df.tmdb_score
```

```python
plt.figure(figsize=(15,7))
plt.xticks(rotation=80)
plt.xlabel("RATINGS")
plt.ylabel('MOVIES')
plt.plot(TITLE1,im12,marker='*')
plt.plot(TITLE1,tm12,marker='o')
plt.legend(['imdb_votes','tmdb_score'])
```

```python
comment_words = ' '
  
for val in raw.title.values: 
    val = str(val) 
    tokens = val.split() 
        
    for i in range(len(tokens)): 
        tokens[i] = tokens[i].lower() 
          
    for words in tokens: 
        comment_words = comment_words + words + ' '
  
  
wordcloud = WordCloud(width = 1000, height = 1000, 
                background_color ='black', 
                min_font_size = 10).generate(comment_words) 
wordcloud.to_image()
```
