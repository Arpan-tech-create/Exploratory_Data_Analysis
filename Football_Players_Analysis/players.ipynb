# Analysis On Argentina Team

```python
player=raw[['short_name','potential','value_eur','defending_sliding_tackle','goalkeeping_handling','dribbling']]
player_df=raw.nationality=='Argentina'
```

```python
new_pl=player[player_df]
final_df=new_pl.sort_values(by=['potential'],ascending=False).head(10)
final_df
```

# As per Football Potential

```python
plt.figure(figsize=(15,6))
plt.xticks(rotation=60)

sns.barplot(x='short_name',y='potential',data=final_df)
```

# Goalkeeping & Sliding Tackle

```python
players=final_df.short_name
pot=final_df.goalkeeping_handling
val=final_df.defending_sliding_tackle
```

```python
plt.figure(figsize=(15,6))
plt.xticks(rotation=60)
plt.plot(players,pot,marker='o',linestyle='--')
plt.plot(players,val,marker='o')
plt.legend(['goalkeeping_handling','defending_sliding_tackle'])
```

# Dribbling

```python
plt.figure(figsize=(15,6))
plt.xticks(rotation=60)

sns.barplot(x='short_name',y='dribbling',data=final_df,palette='rocket')
```
