# New_cases in South Americas Region

```python
south=read1[['Country_Region','New_cases','New_deaths','New_recovered']]
south_basic=read1.WHO_Region=='Americas'
```

```python
ame=south[south_basic]
ame_df=ame.sort_values(by=['New_cases'],ascending=False)
ame_df
```

Max Number Of new cases in : US

In Above Data , There Are 0 values in 142 & belows rows , i will clean it

```python
ame_df.drop(ame_df[ame_df.New_cases<=0].index,inplace=True)
```

```python
ame_df
```

```python
plt.title("New-Cases in South America")
plt.ylabel("Country-Reg",size=16)
plt.xlabel('Number of New Cases',size=16)
plt.xticks(rotation=60)
plt.scatter(y='Country_Region',x='New_cases',data=ame_df)
```

```python
comment_words = ' '
  
for val in read1.WHO_Region.values: 
    val = str(val) 
    tokens = val.split() 
        
    for i in range(len(tokens)): 
        tokens[i] = tokens[i].lower() 
          
    for words in tokens: 
        comment_words = comment_words + words + ' '
  
  
wordcloud = WordCloud(width = 1000, height = 1000, 
                background_color ='black', 
                min_font_size = 10).generate(comment_words) 
wordcloud.to_image()
```
