```python
data1=pd.read_csv('FIFA - 2018.csv')
```

```python
data1
```

```python
data1.sort_values(by=['Points'],ascending=False).head(10)
```

```python
data1.drop(data1[data1.Points<9].index,inplace=True)
```

```python
data1
```

```python
data1.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.size'] = 15
plt.figure(figsize=(12,6))
plt.xticks(rotation=60)
plt.title("Team-Points",size=17,color='b')

sns.barplot(x='Team',y='Win',data=data1)
```

```python
plt.figure(figsize=(10,6))
sns.relplot(x='Team',y='Win',data=data1.head())
```

```python
data1.drop(['Position'],axis=1,inplace=True)
```

```python
data1
```

```python
team_wins=data1.groupby('Team')[['Win']].count()
team_wins.sort_values('Win').reset_index()
team_wins.columns=["Number of points"]
team_wins
```

```python
sns.relplot(x='Team',y='Win',data=data1.head(5))
```

```python
sns.relplot(x='Points',y='Win',hue='Team',data=data1)
```

```python
sns.catplot(x="Team",y="Win",data=data1.head(),kind='box',hue='Points')
```
