```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

plt.rcParams['font.size'] = 13
plt.rcParams['figure.figsize'] = (9, 5)
plt.rcParams['figure.facecolor'] = '#00000000'
```

```python
read1=pd.read_csv("IPL Matches 2008-2020.csv")
read1
```

```python
read1.drop(['neutral_venue'],axis=1,inplace=True)
read1
```

```python
read1.shape
```

```python
read1.drop(['id'],axis=1,inplace=True)
```

```python
read1
```

```python
read1.drop(['date'],axis=1,inplace=True)
```

```python
read1
```

```python
read1.isnull().sum()
```

```python
read1.drop(read1[read1.city.isnull()].index, inplace=True)
read1.drop(read1[read1.player_of_match.isnull()].index, inplace=True)
read1.drop(read1[read1.method.isnull()].index, inplace=True)
```

```python
read1.info()
```

# Number Of Winning Matches Of Team

```python
count=read1['winner'].value_counts()
count
```

```python
plt.figure(figsize=(20,7))
plt.xticks(rotation=70)
plt.title("Most Wins")
sns.barplot(x=count.index,y=count)
```
