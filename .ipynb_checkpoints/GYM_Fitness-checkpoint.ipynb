```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
sns.set_style('darkgrid')
plt.rcParams['font.size']=14
plt.rcParams['figure.figsize']=(13,7)
```

```python
raw_data=pd.read_csv('CardioGoodFitness.csv')
raw_data
```

```python
raw_data.isnull().sum()
```

```python
raw_data.Gender.value_counts()
```

```python
plt.pie(raw_data.Gender.value_counts(),labels=raw_data.Gender.value_counts().index,autopct='%1.1f%%')
plt.title('Gender')
plt.show()
```

```python
male_cnt=raw_data.Gender[raw_data.MaritalStatus=='Single']
male_cnt.value_counts()
```

```python
par_cnt=raw_data.Gender[raw_data.MaritalStatus=='Partnered']
par_cnt.value_counts()
```

```python
raw_data.MaritalStatus.value_counts()
```

```python
sns.pairplot(raw_data,hue='MaritalStatus')
```

```python
px.box(raw_data,x='MaritalStatus',y='Income',color='Gender')
```

```python
plt.xlabel('Income')
plt.ylabel('Miles')
plt.scatter(x='Income',y='Miles',data=raw_data)
```
